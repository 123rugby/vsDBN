```python
from RBM import RBM
import numpy as np

def DBN(filename):
    print(filename)
    data=np.load(filename)
    rbm1=RBM(data.shape[1],100,epochs=20)
    rbm1.train(data)
    h1=rbm1.predict(data)

    rbm2=RBM(100,100,epochs=20)
    rbm2.train(h1)
    h2=rbm2.predict(h1)

    rbm3=RBM(100,100,epochs=20)
    rbm3.train(h2)
    return rbm1.getW(),rbm2.getW(),rbm3.getW()
```

```python
tasks=['EMOTION','GAMBLING','LANGUAGE','MOTOR','RELATIONAL','SOCIAL','WM']
for taskname in tasks:
    filename='/home/share/TmpData/Qinglin/HCP_Group/'+taskname+'100_0.2.npy'
    W1,W2,W3=DBN(filename)
    np.save('/home/share/TmpData/Qinglin/HCP_Group_DBN/'+taskname+'_W1.npy',W1)
    np.save('/home/share/TmpData/Qinglin/HCP_Group_DBN/'+taskname+'_W2.npy',W2)
    np.save('/home/share/TmpData/Qinglin/HCP_Group_DBN/'+taskname+'_W3.npy',W3)
```
