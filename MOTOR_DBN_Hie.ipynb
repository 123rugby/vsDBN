```python
import numpy as np
def MapHierach( a, b):
    I = np.count_nonzero(np.logical_and(a, b))
    A=np.count_nonzero(a)
    B=np.count_nonzero(b)
    return I/min([A,B])

```

```python
from sklearn import metrics
def LayerHierach(A,B):
    D=[]
    for i,a in enumerate(A):
        D.append([])
        print(i)
        for j,b in enumerate(B):
            score=MapHierach(a,b)
            D[i].append(score)
    return D
```
